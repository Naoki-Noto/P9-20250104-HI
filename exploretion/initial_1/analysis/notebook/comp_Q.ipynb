```python
import pandas as pd
from collections import Counter
```

```python
# データの全体リスト
All_data_names = ['Reaction_CO_1.5h', 'Reaction_CO_7.5h', 'Reaction_CO_biphenyl', 'Reaction_CO_ortho', 'Reaction_CO_Cl', 'Reaction_CS', 
                  'Reaction_CN', 'Reaction_2+2', 'Reaction_CF3', 'Reaction_CH2CF3', 'Reaction_CH2F', 'Reaction_Cy', 'Reaction_OCH2F', 'Reaction_SCF3',
                  'Reaction_Masuda', 'Reaction_Sumida']

comparison_dict = {
    "first_select_avarage": 0,
    "second_select_average": 1,
    "third_select_average": 2,
    "fourth_select_average": 3,
    "last_select_average": -1,
}

# 結果を格納する辞書
result_tables = {}

for target in All_data_names:
    
    print("~*~" * 15)
    print(f"{target}")
    
    # 入力辞書
    data_names_dict = {
        #'TrAdaBoostR2_Q_LGBM_ne5_sourcetop5': f'TrAdaBoostR2_Q_LGBM_ne5_{target}_sourcetop5_result',
        #'TrAdaBoostR2_Q_ep0.05_LGBM_ne5_sourcetop5_result': f'TrAdaBoostR2_Q_ep0.05_LGBM_ne5_{target}_sourcetop5_result',
        #'TrAdaBoostR2_Q_ep0.07_LGBM_ne5_sourcetop5_result': f'TrAdaBoostR2_Q_ep0.07_LGBM_ne5_{target}_sourcetop5_result',
        #'TrAdaBoostR2_Q_ep0.1_LGBM_ne5_sourcetop5_result': f'TrAdaBoostR2_Q_ep0.1_LGBM_ne5_{target}_sourcetop5_result',
        #'TrAdaBoostR2_Qtop3_LGBM_ne5_sourcetop5_result': f'TrAdaBoostR2_Qtop3_LGBM_ne5_{target}_sourcetop5_result',
        #'TrAdaBoostR2_Qtop1_LGBM_ne5_sourcetop5_result': f'TrAdaBoostR2_Qtop1_LGBM_ne5_{target}_sourcetop5_result',
        #'TrAdaBoostR2_QAll_LGBM_ne5_sourcetop5_result': f'TrAdaBoostR2_QAll_LGBM_ne5_{target}_sourcetop5_result',
        'XGBRegression_Qtop5_result': f'XGBRegression_Qtop5_{target}_result',
        'RFRegression_Qtop5_result': f'RFRegression_Qtop5_{target}_result',
        'LinearRegression_Qtop5_standard_result': f'LinearRegression_Qtop5_{target}_standard_result',
        #'BayesianOptimization_GPyOpt_EI_RBF': f'BayesianOptimization_{target}_GPyOpt_EI_RBF_result',
}
    inverse_dict = {v: k for k, v in data_names_dict.items()}

    # データを格納する辞書
    data_frames = {}
    
    # ファイルの読み込みと変換
    for name in data_names_dict.values():
        file_path = f"../results/{name}_select_averages.csv"  # ファイルパス
        df = pd.read_csv(file_path)  # データフレームを読み込む
        df.columns = [name]  # 列名をデータフレーム名に変更
        data_frames[name] = df  # 辞書に保存
        
    comparison_table = pd.concat(data_frames.values(), axis=1)
    
    # 指定行の最小値を持つ列を複数取得する
    for key, value in comparison_dict.items():
        row_values = comparison_table.iloc[value]  # 指定行の値を取得
        min_value = row_values.min()  # 最小値を取得
        min_columns = row_values[row_values == min_value].index.tolist()  # 最小値を持つ列名のリスト

        # 最小値の列名に対応するキーを取得
        corresponding_keys = [inverse_dict[col] for col in min_columns if col in inverse_dict]
        
        print(f"The minimum value in {key}: {corresponding_keys}")

        # target に対応する辞書がすでにある場合は更新、なければ新規作成
        if f'{target}' not in result_tables:
            result_tables[f'{target}'] = {}  # 新しい辞書を作成        
        # 結果を追加
        result_tables[f'{target}'].update({f'{key}': corresponding_keys})
        
print("~*~" * 15)

for key_name in comparison_dict.keys():
    specific_key_counter = Counter()
    
    # 特定のキーだけを集計
    for outer_key, inner_dict in result_tables.items():
        for key, values in inner_dict.items():
            if key == f"{key_name}":  # 集計対象のキー
                specific_key_counter.update(values)
    
    # 結果を表示
    print(f"{key_name}_top:")
    for element, count in specific_key_counter.items():
        print(f"{element}: {count}")

    print()       
```
