```python
import pandas as pd
import ast
import numpy as np
```

```python
## Main results ##
```

```python
print('== BO_UCB ==')
All_data_names = ['Reaction_OCH2F', 'Reaction_P', 'Reaction_Si']
for data in All_data_names:
    df = pd.read_csv(f"../results/BOsearch_UCB/BOsearch_{data}_result.csv")
    step_indices = []
    for row in df['actual_rewards_per_step']:
        rewards = ast.literal_eval(row)
        max_val = max(rewards)
        first_max_index = rewards.index(max_val) + 1
        step_indices.append(first_max_index)
    average_step_to_best = np.mean(step_indices)
    std_step_to_best = np.std(step_indices)

    print(f"{data}: avg={average_step_to_best:.2f}, std={std_step_to_best:.2f}")
```
