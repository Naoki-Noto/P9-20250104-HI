```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
print("================================")
print("Descriptor: DFT")
print()

model_list = ["RFsearch", "TrABsearch_all_e0_ne5"]
model_colors = {"RFsearch": "blue", "TrABsearch_all_e0_ne5": "red"}
model_labels = {"RFsearch": "RF", "TrABsearch_all_e0_ne5": "TrAB"}

plt.figure(figsize=(10, 5))

for model in model_list:
    if model == 'TrABsearch_all_e0_ne5':
        file_name = 'TrABsearch_Reaction_CO_biphenyl_top14_e0_ne5_result'
    elif model == 'RFsearch':
        file_name = 'RFsearch_Reaction_CO_biphenyl_result'
    else:
        print(f'Not acceptable: {model}')
        continue

    stats_df = pd.read_csv(f'./results/summary_Oos/{model}/{file_name}_OoS_stats.csv')
    summary_df = pd.read_csv(f'./results/summary_Oos/{model}/{file_name}_OoS_summary.csv')

    if model == 'TrABsearch_all_e0_ne5':
        print(f"Model: TrAB")
    else:
        print(f"Model: RF")
    print(stats_df)
    print()
  
    plt.plot(summary_df['Run_index'], summary_df['total_selected_OPS'], label=model_labels[model], marker='o', color=model_colors[model])

print("================================")
plt.xlabel("Run Index")
plt.ylabel("Selected OPS Count")
plt.xlim(-2,61)
plt.ylim(0,60)
plt.legend()
plt.savefig('./figures/OPS61_DFT.pdf')
plt.show()
```
