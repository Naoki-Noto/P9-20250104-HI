```python
import os
import pandas as pd
from tabulate import tabulate
from itertools import product
```

```python
model_list = [
    "TrAdaBoost",
    #"XGBRegression",
    #"RFRegression",
    #"LinearRegression_standard",
    #"BayesianOptimization",
]
es_list = [
    #"XGB", 
    "LGBM",
]
ne_list = [
    1,
    3,
    5,
    10,
]
k_list = [
    #1, 
    #3,
    #5,
    15,
]
selection_order_list = [
    "top",
    #"bottom",
]
top_k_list = [
    #1,
    3,
    #5,
    #10,
]
epsilon_list = [
    0,
    #0.05,
    #0.1,
    ]
acquisition_type_list = [
    "EI",
    "MPI",
    ]
target = 'Reaction_2+2'
"""
All_data_names = ['Reaction_CO_1.5h', 'Reaction_CO_7.5h', 'Reaction_CO_biphenyl', 'Reaction_CO_ortho', 'Reaction_CO_Cl', 'Reaction_CS', 
                  'Reaction_CN', 'Reaction_2+2', 'Reaction_CF3', 'Reaction_CH2CF3', 'Reaction_CH2F', 'Reaction_Cy', 'Reaction_OCH2F', 'Reaction_SCF3',
                 'Reaction_Masuda', 'Reaction_Sumida']
"""
```

```python
# 入力リスト作成（_select_averages.csvを除いた名前）
files_to_check = []

for model in model_list:
    if model == "TrAdaBoost":
        for es, ne, k, selection_order, top_k, epsilon in product(
            es_list, ne_list, k_list, selection_order_list, top_k_list, epsilon_list
        ):
            files_to_check.append(f'TrAdaBoostR2_Qtop{top_k}_ep{epsilon}_{es}_ne{ne}_{target}_source{selection_order}{k}_result')
    elif model == "BayesianOptimization":
        for acquisition_type in acquisition_type_list:
            files_to_check.append(f'BayesianOptimization_{target}_GPyOpt_{acquisition_type}_RBF_result')
    else:
        files_to_check.append(f'{model}_Qtop{top_k}_ep{epsilon}_{target}_result')

# 存在チェック
file_status = [{"ファイル名": file, "存在": "✔ ある" if os.path.exists(f"../../results/{file}.csv") else "✖ ない"} for file in files_to_check]

# DataFrameに変換
df = pd.DataFrame(file_status)

# 表示
print(tabulate(df, headers="keys", tablefmt="grid"))
```
