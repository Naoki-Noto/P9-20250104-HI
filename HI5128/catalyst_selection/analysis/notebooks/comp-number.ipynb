```python
import pandas as pd
from collections import Counter
from itertools import product
import matplotlib.pyplot as plt
import os
```

```python
target = 'Reaction_CO_Cl'
"""
All_data_names = ['Reaction_CO_1.5h', 'Reaction_CO_7.5h', 'Reaction_CO_biphenyl', 'Reaction_CO_ortho', 'Reaction_CO_Cl', 'Reaction_CS', 
                  'Reaction_CN', 'Reaction_2+2', 'Reaction_CF3', 'Reaction_CH2CF3', 'Reaction_CH2F', 'Reaction_Cy', 'Reaction_OCH2F', 'Reaction_SCF3',
                 'Reaction_Masuda', 'Reaction_Sumida']
"""

model_list = [
    "TrAdaBoost",
    #"XGBRegression",
    #"RFRegression",
    #"LinearRegression_standard",
]
es_list = [
    #"XGB", 
    "LGBM",
]
ne_list = [
    #1,
    #3,
    5,
    #10,
]
k_list = [
    #1, 
    #3,
    5,
    15,
]
selection_order_list = [
    "top",
    "bottom",
]
top_k_list = [
    1,
    #3,
    #5,
    #10,
]
epsilon_list = [
    0,
    #0.05,
    #0.1,
    ]
acquisition_type_list = [
    #"EI",
    #"MPI",
    ]
```

```python
comparison_dict = {
    "first_select_avarage": 0,
    "second_select_average": 1,
    "third_select_average": 2,
    "fourth_select_average": 3,
    "last_select_average": -1,
}

# 結果を格納する辞書
result_tables = {}

print("~*~" * 15)
print(f"{target}")

# 入力辞書
data_names_dict = {}
for model in model_list:
    if model == "TrAdaBoost":
        for es, ne, k, selection_order, top_k, epsilon in product(
            es_list, ne_list, k_list, selection_order_list, top_k_list, epsilon_list
        ):
            data_names_dict[f'TrAdaBoostR2_Qtop{top_k}_ep{epsilon}_{es}_ne{ne}_source{selection_order}{k}'] = f'TrAdaBoostR2_Qtop{top_k}_ep{epsilon}_{es}_ne{ne}_{target}_source{selection_order}{k}_result'
    elif model == "BayesianOptimization":
        for acquisition_type in acquisition_type_list:
            data_names_dict[f'BayesianOptimization_GPyOpt_{acquisition_type}_RBF'] = f'BayesianOptimization_{target}_GPyOpt_{acquisition_type}_RBF_result'
    else:
        data_names_dict[f'{model}_Qtop{top_k}_ep{epsilon}_result'] = f'{model}_Qtop{top_k}_ep{epsilon}_{target}_result'
inverse_dict = {v: k for k, v in data_names_dict.items()}

# データを格納する辞書
data_frames = {}

# ファイルの読み込みと変換
for name in data_names_dict.values():
    file_path = f"../results/{name}_select_averages.csv"  # ファイルパス
    if os.path.exists(file_path):  # ファイルが存在する場合のみ読み込む
        df = pd.read_csv(file_path)  # データフレームを読み込む
        df.columns = [name]  # 列名をデータフレーム名に変更
        data_frames[name] = df  # 辞書に保存
    
comparison_table = pd.concat(data_frames.values(), axis=1)

display(comparison_table)
        
print("~*~" * 15)
```
