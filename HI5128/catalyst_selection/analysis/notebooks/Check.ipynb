```python
import pandas as pd
```

```python
data_names = [
    "TrAdaBoostR2_Qtop3_ep0_XGB_ne5_Reaction_2+2_sourcetop15",
    "BayesianOptimization_Reaction_2+2_GPyOpt_EI_RBF",
    "LinearRegression_standard_Qtop3_ep0_Reaction_2+2",
    "XGBRegression_Qtop3_ep0_Reaction_2+2",
    "RFRegression_Qtop3_ep0_Reaction_2+2",
]

for model in data_names:
    print("=" * 20)
    print(f"{model}")
    df1 = pd.read_csv(f"../../results/{model}_result.csv")
    df2 = pd.read_csv(f"../../results/{model}_check_result.csv")

    for i in range(60):
        index = i
        
        df1['initial_indices'] = df1['initial_indices'].astype(str).str.strip("[]").astype(int)
        df2['initial_indices'] = df2['initial_indices'].astype(str).str.strip("[]").astype(int)
        
        result1 = df1.loc[df1['initial_indices'] == index, 'selected_indices_per_step'].values
        result2 = df2.loc[df2['initial_indices'] == index, 'selected_indices_per_step'].values

        if result1 == result2:
            print(f"index {i} : OK")

        else:
            print(f"index {i} : No match")
            
```
