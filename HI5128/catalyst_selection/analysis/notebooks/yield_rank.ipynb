```python
import pandas as pd
```

```python
target_index = 0

All_data_names = ['Reaction_CO_1.5h', 'Reaction_CO_7.5h', 'Reaction_CO_biphenyl', 'Reaction_CO_ortho', 'Reaction_CO_Cl', 'Reaction_CS', 
                  'Reaction_CN', 'Reaction_2+2', 'Reaction_CF3', 'Reaction_CH2CF3', 'Reaction_CH2F', 'Reaction_Cy', 'Reaction_OCH2F', 'Reaction_SCF3',
                 'Reaction_Masuda', 'Reaction_Sumida']

original_data_dict = {}
for target in All_data_names:
    original_data_dict[target] = pd.read_csv(f'../../../data/data_60_{target}.csv')

sorted_data_dict = {}
for target in All_data_names:
    df = original_data_dict[target]
    sorted_df = df.sort_values(by="Yield", ascending=False)
    
    # 順位を振る
    sorted_df["Rank"] = sorted_df["Yield"].rank(method="min", ascending=False).astype(int)

    # 辞書に入れる
    sorted_data_dict[target] = sorted_df
    
print(f"index {target_index} is")
for target, sorted_df in sorted_data_dict.items():
    print(f"Rank {sorted_df[["Rank"]].loc[target_index].item()} in {target}")

for target, sorted_df in sorted_data_dict.items():
    # 結果を表示
    print("=" * 30)
    print(target)
    print(sorted_df[["Rank", "Name", "ID", "Yield"]])
```
